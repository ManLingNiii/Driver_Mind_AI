```python
# 區塊 5: 模型訓練：定義 YOLOv8 訓練核心函式
# ------------------------------------------------------------------------------
# 從 ultralytics 函式庫中引入 YOLO 類別 (此類別在區塊 3 已引入，重複引入無害)
from ultralytics import YOLO
import os # 引入 os 模組 (已引入，重複引入無害)

print("定義 YOLOv8 訓練函式 train_yolov8...")

def train_yolov8(
    model_path: str,          # 基礎模型權重路徑 (例如: 'yolov8n.pt', 'yolov8m.pt', 或前一次訓練的 'best.pt')
    yaml_path: str,           # 資料集設定檔 (.yaml) 路徑 (參閱區塊 4)
    mode: str,             # 訓練模式：'default', 'simple', 'custom', 'tune'
    project_name: str,         # 訓練結果儲存於 runs/detect/project_name
    epochs: int,            # 訓練輪數
    batch: int,             # 批次大小 (請注意顯存限制)
    imgsz: int =1280,         # 圖片尺寸 (建議是 32 的倍數，且與您的資料集尺寸相符)
    optimizer: str = "sgd",  # 優化器 (如 'SGD', 'Adam', 'AdamW', 或 'auto' 由 YOLOv8 自動選擇)
    # 其他 YOLOv8 train 參數可以在 model.train() 中添加
):
    """
    YOLOv8 模型訓練核心函式。

    Args:
        model_path (str): 基礎模型權重路徑。
        yaml_path (str): 資料集設定檔 (.yaml) 路徑。
        mode (str): 訓練模式 ('default', 'simple', 'custom', 'tune')。
        project_name (str): 訓練結果在 runs/detect/ 下建立的資料夾名稱。
        epochs (int): 訓練輪數。
        batch (int): 訓練批次大小。
        imgsz (int): 訓練圖片尺寸。
        optimizer (str): 訓練優化器，用於 tune 模式。
        # hyp (str, optional): 指定超參數檔案路徑。預設 None。
    """

    print(f"\n--- 初始化模型：從 {model_path} 載入權重 ---")
    try:
        model = YOLO(model_path)
        print("✅ 模型初始化成功。")
    except Exception as e:
        print(f"❌ 模型初始化失敗：{e}")
        print("請檢查 model_path 是否正確且模型檔案存在。")
        return


    print(f"--- 🛠️ 開始執行訓練模式: {mode} ---")
    print(f"訓練參數: Epochs={epochs}, Batch Size={batch}, Image Size={imgsz}, Project Name={project_name}")
    if mode != "tune":
         print(f"資料集設定檔: {yaml_path}")


    # 根據選擇的模式執行訓練
    if mode == "default":
        # 使用 YOLOv8 的預設訓練參數
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_default", # 訓練結果資料夾名稱
            verbose=True                    # 顯示詳細訓練過程
        )

    elif mode == "simple":
        # 使用一組簡化的自定義超參數
        # 注意：這些參數會覆蓋預設值
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_simple",
            lr0=0.005,                 # 初始學習率（比預設的 0.01 保守一點，較穩定）
            weight_decay=0.0005,            # 權重衰減（防止 overfitting，控制模型複雜度）
            flipud=0.0,                # 垂直翻轉的機率（設定為 0：關閉，因為車道資料不適合垂直翻轉）
            fliplr=0.5,                # 水平翻轉的機率（設定為 0.5，可模擬左右對稱的場景）
            mosaic=1.0,                # Mosaic 資料增強（開啟能提升小樣本表現）
            mixup=0.0,                # MixUp 資料增強（小樣本不建議開，可能使模型混淆）
            verbose=True               # 顯示詳細訓練 log（每個 batch 輸出 loss）
        )




    elif mode == "custom":
        # 使用一組更完整的自定義超參數
        print(f"✅ 使用 custom 模式，並套用進階自定義超參數")
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_custom",

            # === Early Stopping ===
            patience=30,

            # === 超參數（學習率相關）===
            lr0=0.00745,
            lrf=0.00896,
            momentum=0.833,
            weight_decay=0.00049,
            warmup_epochs=4.65213,
            warmup_momentum=0.84891,

            # === 損失函數權重 ===
            box=4.92998,
            cls=0.59779,
            dfl=1.69388,

            # === 資料增強參數 ===
            hsv_h=0.01749,
            hsv_s=0.58542,
            hsv_v=0.26096,
            degrees=0.0,
            translate=0.14309,
            scale=0.4027,
            shear=0.0,
            perspective=0.0,
            flipud=0.0,
            fliplr=0.35866,
            bgr=0.0,
            mosaic=0.70593,
            cutmix=0.0,
            mixup=0,
            copy_paste=0,

            # === 其他 ===
            optimizer="auto",
            verbose=True,

        )





    elif mode == "tune":
        # 使用 tune 模式進行超參數搜尋 (Hyperspace Evolution)
        # 這會執行多次實驗，尋找最佳的超參數組合
        # iterations 參數控制搜尋的實驗次數
        iterations = 20 # 您可以依照需要調整搜尋迭代次數
        print(f"執行超參數搜尋 (Hyperspace Evolution)，共 {iterations} 次迭代。")

        model.tune(
            data=yaml_path,
            epochs=10,       # 每次實驗的訓練輪數
            imgsz=1280,
            iterations=iterations, # 搜尋的總迭代次數
            optimizer="sgd", # 用於 tune 的優化器
            plots=True,          # 產生訓練圖表
            name=f"{project_name}_tune" # 超參數搜尋結果的資料夾名稱
        )
        print(f"超參數搜尋完成。結果儲存於 runs/detect/{project_name}_tune")

    else:
        # 處理無效的 mode 選項
        print("\n❌ 無效的 mode 選項。請使用：'default', 'simple', 'custom', 'tune'")
        return

    print("\n--- 訓練過程已啟動（請查看上方輸出確認狀態）---")
    print("訓練完成後，結果通常會儲存於 ./runs/detect/ 下的對應資料夾。")

print("\nYOLOv8 訓練函式 train_yolov8 定義完成。")
```

```python
# 區塊 6: 模型訓練：執行訓練函式
# ------------------------------------------------------------------------------

print("--- 執行 YOLOv8 模型訓練 ---")

# 呼叫前面定義的 train_yolov8 函式，並傳入訓練參數
# 請根據您的需求修改以下參數：

train_yolov8(
    # 基礎模型：可以選擇 yolov8n.pt (nano, 最小最快), yolov8s.pt (small),
    #          yolov8m.pt (medium), yolov8l.pt (large), yolov8x.pt (extra large)
    #          或者使用自己之前訓練好的模型權重路徑 (例如: 'runs/detect/previous_run/weights/best.pt')
    model_path="yolov8s.pt",

    # 資料集設定檔路徑：請確保與區塊 4 設定的 yaml_path 一致
    yaml_path=yaml_path, # 直接使用前面設定的 yaml_path 變數

    # 訓練模式：選擇您想使用的訓練模式
    # 'default': 使用 YOLOv8 內建預設參數
    # 'simple': 使用 train_yolov8 函式中定義的一組簡單自定義參數
    # 'custom': 使用 train_yolov8 函式中定義的一組更詳細的自定義參數
    # 'tune': 執行超參數搜尋 (會進行多次實驗)
    mode="custom", # <<< 修改此處選擇訓練模式

    # 訓練結果專案名稱：訓練結果將儲存於 ./runs/detect/ 這個名稱的資料夾下
    project_name="2421_1280*1280_200epochs40_s_tune", # <<< 修改此處設定專案名稱

    # 訓練輪數：模型將完整跑過訓練資料集的次數
    epochs=200, # <<< 設定訓練輪數

    # 批次大小：每次梯度更新使用的圖片數量。請根據您的 GPU 顯存調整。
    # 如果出現 CUDA 記憶體不足錯誤，請降低 batch 大小。
    batch=16, # <<< 設定批次大小

    # 圖片尺寸：訓練時圖片會被縮放或填充到此尺寸。應是 32 的倍數。
    imgsz=1280, # <<< 設定圖片尺寸 (與資料集或預測尺寸相符通常較好)

    # 優化器：主要用於 tune 模式，其他模式通常由模型預設決定
    optimizer="sgd" # 例如: 'SGD', 'Adam', 'AdamW'

    # 如果是 'tune' 模式，iterations 會在函式內部設定 (預設 20 次)
    # 如果是 'custom' 或 'simple' 模式，內建的超參數會生效
    # 如果您想使用外部超參數檔案，可以在 train_yolov8 函式定義中取消 hyp 參數的註解並在此傳入路徑
    patience=30,  # <<< 設定 Early Stopping 的耐心值：若連續 30 次 epoch val_loss 沒有改善，則提前停止訓練
)

print("\n--- train_yolov8 函式呼叫完成。訓練已開始或即將開始 (請查看上方輸出)。---")
```

```python
# 區塊 7: 結果處理：備份訓練結果至 Google Drive
# ------------------------------------------------------------------------------
import shutil # 引入 shutil 模組，用於高階檔案操作 (如複製整個資料夾)
import os     # 引入 os 模組 (已引入，重複引入無害)
from google.colab import drive

# 調用 drive.mount() 函式來掛載 Google Drive
# 指定掛載點為 '/content/drive'
# 執行後會要求進行 Google 帳號授權
print("請依照指示進行 Google Drive 授權...")
drive.mount('/content/drive')

print("--- 開始備份訓練結果至 Google Drive ---")

# 步驟 7.1: 指定原始資料夾路徑 (YOLOv8 訓練結果預設儲存位置)
# 請確認這個路徑是您訓練結果所在的目錄
src_folder = "/content/runs" # 通常是 /content/runs，因為我們將工作目錄設定為 /content/
print(f"原始訓練結果路徑：{src_folder}")

# 步驟 7.2: 指定 Google Drive 中的備份目標路徑
# 請修改此處以指定您希望備份結果到 Drive 的哪個位置
# 例如：備份到 '我的雲端硬碟' 下的 'yolov8_backups' 資料夾
backup_folder = "/content/drive/My Drive/runs_0605_2421images_1280*1280_s_good_backup/" # <<< 修改此處設定您的備份路徑
print(f"Google Drive 備份目標路徑：{backup_folder}")


# 步驟 7.3: 檢查原始資料夾是否存在
if not os.path.exists(src_folder):
    print(f"\n⚠️ 警告：原始訓練結果資料夾不存在！無法執行備份：{src_folder}")
    print("請確認訓練是否成功完成，以及 src_folder 路徑是否正確。")
else:
    # 步驟 7.4: 建立備份目標資料夾（如果還不存在）
    # 使用 os.makedirs() 可以建立多層級的資料夾
    print(f"\n檢查或建立備份目標資料夾：{backup_folder}")
    os.makedirs(backup_folder, exist_ok=True) # exist_ok=True 避免資料夾已存在時報錯
    print("備份目標資料夾已確認存在。")

    # 步驟 7.5: 複製整個 runs 資料夾到 Google Drive
    # shutil.copytree() 會遞歸複製整個資料夾樹，包括所有子資料夾和檔案
    # dirs_exist_ok=True 允許目標資料夾已存在，並會嘗試合併內容
    print(f"\n開始複製 {src_folder} 到 {backup_folder}...")
    try:
        shutil.copytree(src_folder, backup_folder, dirs_exist_ok=True)
        print("\n✅ 已成功備份訓練結果至 Google Drive！")
        print(f"備份位置：{backup_folder}")
    except Exception as e:
        print(f"\n❌ 備份失敗！發生錯誤：{e}")
        print("請檢查您的 Google Drive 空間是否充足，或路徑是否有問題。")


print("\n備份流程執行完畢。")
```
