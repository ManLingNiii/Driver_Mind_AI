```python
# å€å¡Š 5: æ¨¡å‹è¨“ç·´ï¼šå®šç¾© YOLOv8 è¨“ç·´æ ¸å¿ƒå‡½å¼
# ------------------------------------------------------------------------------
# å¾ ultralytics å‡½å¼åº«ä¸­å¼•å…¥ YOLO é¡åˆ¥ (æ­¤é¡åˆ¥åœ¨å€å¡Š 3 å·²å¼•å…¥ï¼Œé‡è¤‡å¼•å…¥ç„¡å®³)
from ultralytics import YOLO
import os # å¼•å…¥ os æ¨¡çµ„ (å·²å¼•å…¥ï¼Œé‡è¤‡å¼•å…¥ç„¡å®³)

print("å®šç¾© YOLOv8 è¨“ç·´å‡½å¼ train_yolov8...")

def train_yolov8(
    model_path: str,          # åŸºç¤æ¨¡å‹æ¬Šé‡è·¯å¾‘ (ä¾‹å¦‚: 'yolov8n.pt', 'yolov8m.pt', æˆ–å‰ä¸€æ¬¡è¨“ç·´çš„ 'best.pt')
    yaml_path: str,           # è³‡æ–™é›†è¨­å®šæª” (.yaml) è·¯å¾‘ (åƒé–±å€å¡Š 4)
    mode: str,             # è¨“ç·´æ¨¡å¼ï¼š'default', 'simple', 'custom', 'tune'
    project_name: str,         # è¨“ç·´çµæœå„²å­˜æ–¼ runs/detect/project_name
    epochs: int,            # è¨“ç·´è¼ªæ•¸
    batch: int,             # æ‰¹æ¬¡å¤§å° (è«‹æ³¨æ„é¡¯å­˜é™åˆ¶)
    imgsz: int =1280,         # åœ–ç‰‡å°ºå¯¸ (å»ºè­°æ˜¯ 32 çš„å€æ•¸ï¼Œä¸”èˆ‡æ‚¨çš„è³‡æ–™é›†å°ºå¯¸ç›¸ç¬¦)
    optimizer: str = "sgd",  # å„ªåŒ–å™¨ (å¦‚ 'SGD', 'Adam', 'AdamW', æˆ– 'auto' ç”± YOLOv8 è‡ªå‹•é¸æ“‡)
    # å…¶ä»– YOLOv8 train åƒæ•¸å¯ä»¥åœ¨ model.train() ä¸­æ·»åŠ 
):
    """
    YOLOv8 æ¨¡å‹è¨“ç·´æ ¸å¿ƒå‡½å¼ã€‚

    Args:
        model_path (str): åŸºç¤æ¨¡å‹æ¬Šé‡è·¯å¾‘ã€‚
        yaml_path (str): è³‡æ–™é›†è¨­å®šæª” (.yaml) è·¯å¾‘ã€‚
        mode (str): è¨“ç·´æ¨¡å¼ ('default', 'simple', 'custom', 'tune')ã€‚
        project_name (str): è¨“ç·´çµæœåœ¨ runs/detect/ ä¸‹å»ºç«‹çš„è³‡æ–™å¤¾åç¨±ã€‚
        epochs (int): è¨“ç·´è¼ªæ•¸ã€‚
        batch (int): è¨“ç·´æ‰¹æ¬¡å¤§å°ã€‚
        imgsz (int): è¨“ç·´åœ–ç‰‡å°ºå¯¸ã€‚
        optimizer (str): è¨“ç·´å„ªåŒ–å™¨ï¼Œç”¨æ–¼ tune æ¨¡å¼ã€‚
        # hyp (str, optional): æŒ‡å®šè¶…åƒæ•¸æª”æ¡ˆè·¯å¾‘ã€‚é è¨­ Noneã€‚
    """

    print(f"\n--- åˆå§‹åŒ–æ¨¡å‹ï¼šå¾ {model_path} è¼‰å…¥æ¬Šé‡ ---")
    try:
        model = YOLO(model_path)
        print("âœ… æ¨¡å‹åˆå§‹åŒ–æˆåŠŸã€‚")
    except Exception as e:
        print(f"âŒ æ¨¡å‹åˆå§‹åŒ–å¤±æ•—ï¼š{e}")
        print("è«‹æª¢æŸ¥ model_path æ˜¯å¦æ­£ç¢ºä¸”æ¨¡å‹æª”æ¡ˆå­˜åœ¨ã€‚")
        return


    print(f"--- ğŸ› ï¸ é–‹å§‹åŸ·è¡Œè¨“ç·´æ¨¡å¼: {mode} ---")
    print(f"è¨“ç·´åƒæ•¸: Epochs={epochs}, Batch Size={batch}, Image Size={imgsz}, Project Name={project_name}")
    if mode != "tune":
         print(f"è³‡æ–™é›†è¨­å®šæª”: {yaml_path}")


    # æ ¹æ“šé¸æ“‡çš„æ¨¡å¼åŸ·è¡Œè¨“ç·´
    if mode == "default":
        # ä½¿ç”¨ YOLOv8 çš„é è¨­è¨“ç·´åƒæ•¸
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_default", # è¨“ç·´çµæœè³‡æ–™å¤¾åç¨±
            verbose=True                    # é¡¯ç¤ºè©³ç´°è¨“ç·´éç¨‹
        )

    elif mode == "simple":
        # ä½¿ç”¨ä¸€çµ„ç°¡åŒ–çš„è‡ªå®šç¾©è¶…åƒæ•¸
        # æ³¨æ„ï¼šé€™äº›åƒæ•¸æœƒè¦†è“‹é è¨­å€¼
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_simple",
            lr0=0.005,                 # åˆå§‹å­¸ç¿’ç‡ï¼ˆæ¯”é è¨­çš„ 0.01 ä¿å®ˆä¸€é»ï¼Œè¼ƒç©©å®šï¼‰
            weight_decay=0.0005,            # æ¬Šé‡è¡°æ¸›ï¼ˆé˜²æ­¢ overfittingï¼Œæ§åˆ¶æ¨¡å‹è¤‡é›œåº¦ï¼‰
            flipud=0.0,                # å‚ç›´ç¿»è½‰çš„æ©Ÿç‡ï¼ˆè¨­å®šç‚º 0ï¼šé—œé–‰ï¼Œå› ç‚ºè»Šé“è³‡æ–™ä¸é©åˆå‚ç›´ç¿»è½‰ï¼‰
            fliplr=0.5,                # æ°´å¹³ç¿»è½‰çš„æ©Ÿç‡ï¼ˆè¨­å®šç‚º 0.5ï¼Œå¯æ¨¡æ“¬å·¦å³å°ç¨±çš„å ´æ™¯ï¼‰
            mosaic=1.0,                # Mosaic è³‡æ–™å¢å¼·ï¼ˆé–‹å•Ÿèƒ½æå‡å°æ¨£æœ¬è¡¨ç¾ï¼‰
            mixup=0.0,                # MixUp è³‡æ–™å¢å¼·ï¼ˆå°æ¨£æœ¬ä¸å»ºè­°é–‹ï¼Œå¯èƒ½ä½¿æ¨¡å‹æ··æ·†ï¼‰
            verbose=True               # é¡¯ç¤ºè©³ç´°è¨“ç·´ logï¼ˆæ¯å€‹ batch è¼¸å‡º lossï¼‰
        )




    elif mode == "custom":
        # ä½¿ç”¨ä¸€çµ„æ›´å®Œæ•´çš„è‡ªå®šç¾©è¶…åƒæ•¸
        print(f"âœ… ä½¿ç”¨ custom æ¨¡å¼ï¼Œä¸¦å¥—ç”¨é€²éšè‡ªå®šç¾©è¶…åƒæ•¸")
        model.train(
            data=yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            name=f"{project_name}_custom",

            # === Early Stopping ===
            patience=30,

            # === è¶…åƒæ•¸ï¼ˆå­¸ç¿’ç‡ç›¸é—œï¼‰===
            lr0=0.00745,
            lrf=0.00896,
            momentum=0.833,
            weight_decay=0.00049,
            warmup_epochs=4.65213,
            warmup_momentum=0.84891,

            # === æå¤±å‡½æ•¸æ¬Šé‡ ===
            box=4.92998,
            cls=0.59779,
            dfl=1.69388,

            # === è³‡æ–™å¢å¼·åƒæ•¸ ===
            hsv_h=0.01749,
            hsv_s=0.58542,
            hsv_v=0.26096,
            degrees=0.0,
            translate=0.14309,
            scale=0.4027,
            shear=0.0,
            perspective=0.0,
            flipud=0.0,
            fliplr=0.35866,
            bgr=0.0,
            mosaic=0.70593,
            cutmix=0.0,
            mixup=0,
            copy_paste=0,

            # === å…¶ä»– ===
            optimizer="auto",
            verbose=True,

        )





    elif mode == "tune":
        # ä½¿ç”¨ tune æ¨¡å¼é€²è¡Œè¶…åƒæ•¸æœå°‹ (Hyperspace Evolution)
        # é€™æœƒåŸ·è¡Œå¤šæ¬¡å¯¦é©—ï¼Œå°‹æ‰¾æœ€ä½³çš„è¶…åƒæ•¸çµ„åˆ
        # iterations åƒæ•¸æ§åˆ¶æœå°‹çš„å¯¦é©—æ¬¡æ•¸
        iterations = 20 # æ‚¨å¯ä»¥ä¾ç…§éœ€è¦èª¿æ•´æœå°‹è¿­ä»£æ¬¡æ•¸
        print(f"åŸ·è¡Œè¶…åƒæ•¸æœå°‹ (Hyperspace Evolution)ï¼Œå…± {iterations} æ¬¡è¿­ä»£ã€‚")

        model.tune(
            data=yaml_path,
            epochs=10,       # æ¯æ¬¡å¯¦é©—çš„è¨“ç·´è¼ªæ•¸
            imgsz=1280,
            iterations=iterations, # æœå°‹çš„ç¸½è¿­ä»£æ¬¡æ•¸
            optimizer="sgd", # ç”¨æ–¼ tune çš„å„ªåŒ–å™¨
            plots=True,          # ç”¢ç”Ÿè¨“ç·´åœ–è¡¨
            name=f"{project_name}_tune" # è¶…åƒæ•¸æœå°‹çµæœçš„è³‡æ–™å¤¾åç¨±
        )
        print(f"è¶…åƒæ•¸æœå°‹å®Œæˆã€‚çµæœå„²å­˜æ–¼ runs/detect/{project_name}_tune")

    else:
        # è™•ç†ç„¡æ•ˆçš„ mode é¸é …
        print("\nâŒ ç„¡æ•ˆçš„ mode é¸é …ã€‚è«‹ä½¿ç”¨ï¼š'default', 'simple', 'custom', 'tune'")
        return

    print("\n--- è¨“ç·´éç¨‹å·²å•Ÿå‹•ï¼ˆè«‹æŸ¥çœ‹ä¸Šæ–¹è¼¸å‡ºç¢ºèªç‹€æ…‹ï¼‰---")
    print("è¨“ç·´å®Œæˆå¾Œï¼Œçµæœé€šå¸¸æœƒå„²å­˜æ–¼ ./runs/detect/ ä¸‹çš„å°æ‡‰è³‡æ–™å¤¾ã€‚")

print("\nYOLOv8 è¨“ç·´å‡½å¼ train_yolov8 å®šç¾©å®Œæˆã€‚")
```

```python
# å€å¡Š 6: æ¨¡å‹è¨“ç·´ï¼šåŸ·è¡Œè¨“ç·´å‡½å¼
# ------------------------------------------------------------------------------

print("--- åŸ·è¡Œ YOLOv8 æ¨¡å‹è¨“ç·´ ---")

# å‘¼å«å‰é¢å®šç¾©çš„ train_yolov8 å‡½å¼ï¼Œä¸¦å‚³å…¥è¨“ç·´åƒæ•¸
# è«‹æ ¹æ“šæ‚¨çš„éœ€æ±‚ä¿®æ”¹ä»¥ä¸‹åƒæ•¸ï¼š

train_yolov8(
    # åŸºç¤æ¨¡å‹ï¼šå¯ä»¥é¸æ“‡ yolov8n.pt (nano, æœ€å°æœ€å¿«), yolov8s.pt (small),
    #          yolov8m.pt (medium), yolov8l.pt (large), yolov8x.pt (extra large)
    #          æˆ–è€…ä½¿ç”¨è‡ªå·±ä¹‹å‰è¨“ç·´å¥½çš„æ¨¡å‹æ¬Šé‡è·¯å¾‘ (ä¾‹å¦‚: 'runs/detect/previous_run/weights/best.pt')
    model_path="yolov8s.pt",

    # è³‡æ–™é›†è¨­å®šæª”è·¯å¾‘ï¼šè«‹ç¢ºä¿èˆ‡å€å¡Š 4 è¨­å®šçš„ yaml_path ä¸€è‡´
    yaml_path=yaml_path, # ç›´æ¥ä½¿ç”¨å‰é¢è¨­å®šçš„ yaml_path è®Šæ•¸

    # è¨“ç·´æ¨¡å¼ï¼šé¸æ“‡æ‚¨æƒ³ä½¿ç”¨çš„è¨“ç·´æ¨¡å¼
    # 'default': ä½¿ç”¨ YOLOv8 å…§å»ºé è¨­åƒæ•¸
    # 'simple': ä½¿ç”¨ train_yolov8 å‡½å¼ä¸­å®šç¾©çš„ä¸€çµ„ç°¡å–®è‡ªå®šç¾©åƒæ•¸
    # 'custom': ä½¿ç”¨ train_yolov8 å‡½å¼ä¸­å®šç¾©çš„ä¸€çµ„æ›´è©³ç´°çš„è‡ªå®šç¾©åƒæ•¸
    # 'tune': åŸ·è¡Œè¶…åƒæ•¸æœå°‹ (æœƒé€²è¡Œå¤šæ¬¡å¯¦é©—)
    mode="custom", # <<< ä¿®æ”¹æ­¤è™•é¸æ“‡è¨“ç·´æ¨¡å¼

    # è¨“ç·´çµæœå°ˆæ¡ˆåç¨±ï¼šè¨“ç·´çµæœå°‡å„²å­˜æ–¼ ./runs/detect/ é€™å€‹åç¨±çš„è³‡æ–™å¤¾ä¸‹
    project_name="2421_1280*1280_200epochs40_s_tune", # <<< ä¿®æ”¹æ­¤è™•è¨­å®šå°ˆæ¡ˆåç¨±

    # è¨“ç·´è¼ªæ•¸ï¼šæ¨¡å‹å°‡å®Œæ•´è·‘éè¨“ç·´è³‡æ–™é›†çš„æ¬¡æ•¸
    epochs=200, # <<< è¨­å®šè¨“ç·´è¼ªæ•¸

    # æ‰¹æ¬¡å¤§å°ï¼šæ¯æ¬¡æ¢¯åº¦æ›´æ–°ä½¿ç”¨çš„åœ–ç‰‡æ•¸é‡ã€‚è«‹æ ¹æ“šæ‚¨çš„ GPU é¡¯å­˜èª¿æ•´ã€‚
    # å¦‚æœå‡ºç¾ CUDA è¨˜æ†¶é«”ä¸è¶³éŒ¯èª¤ï¼Œè«‹é™ä½ batch å¤§å°ã€‚
    batch=16, # <<< è¨­å®šæ‰¹æ¬¡å¤§å°

    # åœ–ç‰‡å°ºå¯¸ï¼šè¨“ç·´æ™‚åœ–ç‰‡æœƒè¢«ç¸®æ”¾æˆ–å¡«å……åˆ°æ­¤å°ºå¯¸ã€‚æ‡‰æ˜¯ 32 çš„å€æ•¸ã€‚
    imgsz=1280, # <<< è¨­å®šåœ–ç‰‡å°ºå¯¸ (èˆ‡è³‡æ–™é›†æˆ–é æ¸¬å°ºå¯¸ç›¸ç¬¦é€šå¸¸è¼ƒå¥½)

    # å„ªåŒ–å™¨ï¼šä¸»è¦ç”¨æ–¼ tune æ¨¡å¼ï¼Œå…¶ä»–æ¨¡å¼é€šå¸¸ç”±æ¨¡å‹é è¨­æ±ºå®š
    optimizer="sgd" # ä¾‹å¦‚: 'SGD', 'Adam', 'AdamW'

    # å¦‚æœæ˜¯ 'tune' æ¨¡å¼ï¼Œiterations æœƒåœ¨å‡½å¼å…§éƒ¨è¨­å®š (é è¨­ 20 æ¬¡)
    # å¦‚æœæ˜¯ 'custom' æˆ– 'simple' æ¨¡å¼ï¼Œå…§å»ºçš„è¶…åƒæ•¸æœƒç”Ÿæ•ˆ
    # å¦‚æœæ‚¨æƒ³ä½¿ç”¨å¤–éƒ¨è¶…åƒæ•¸æª”æ¡ˆï¼Œå¯ä»¥åœ¨ train_yolov8 å‡½å¼å®šç¾©ä¸­å–æ¶ˆ hyp åƒæ•¸çš„è¨»è§£ä¸¦åœ¨æ­¤å‚³å…¥è·¯å¾‘
    patience=30,  # <<< è¨­å®š Early Stopping çš„è€å¿ƒå€¼ï¼šè‹¥é€£çºŒ 30 æ¬¡ epoch val_loss æ²’æœ‰æ”¹å–„ï¼Œå‰‡æå‰åœæ­¢è¨“ç·´
)

print("\n--- train_yolov8 å‡½å¼å‘¼å«å®Œæˆã€‚è¨“ç·´å·²é–‹å§‹æˆ–å³å°‡é–‹å§‹ (è«‹æŸ¥çœ‹ä¸Šæ–¹è¼¸å‡º)ã€‚---")
```
